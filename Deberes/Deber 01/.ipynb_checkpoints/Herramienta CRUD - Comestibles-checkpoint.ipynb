```python

comestibles = []
class Comestible:
    producto = None
    precio = None
    cantidad = None
    fecha_caducidad = None
    empresa = None
    
    def __init__ (self, producto, precio, cantidad, fecha_caducidad, empresa):
        self.producto = producto
        self.precio = precio
        self.cantidad = cantidad
        self.fecha_caducidad = fecha_caducidad
        self.empresa = empresa
        
    def __str__(self):
        return f"Producto: {self.producto} Precio: {self.precio} Cantidad: {self.cantidad} Fecha de Caducidad:{self.fecha_caducidad} Empresa: {self.empresa}"
    
def nuevo_producto():
        producto = input("Ingresar prodcuto nuevo:\n")
        precio = input ("Ingresar precio del prodcuto:\n")
        cantidad = input("Cantidad a ingresar del producto (paquetes/fundas):\n")
        fecha_caducidad = input("Fecha de caducidad del producto:\n")
        empresa = input("Ingresar proveedor:\n")
        comestibles.append(Comestible(producto, precio, cantidad, fecha_caducidad, empresa))
        print(".....Producto Ingresado.....\n")

def guardar_prodcuto():
    try: 
        path = "./Comestibles.txt"
        archivo_abierto = open(path, 'w')
        guardar=""
        for pro_nuevo in comestibles:
            guardar +=pro_nuevoo.producto+" - "+pro_nuevo.precio+" - "+pro_nuevoro.cantidad+" - "+pro_nuevo.fecha_caducidad+" - "+pro_nuevo.empresa
        archivo_abierto.write(guardar)
        archivo_abierto.close()
    except Exception as Error:
        print("Error al Guardar...!!!")

def imprimir_productos():
    try:
        path = "./Comestibles.txt"
        archivo_abierto = open(path)
        lineas = archivo_abierto.readlines()
        print(lineas)
        archivo_abierto.close()
    except Exception as Error:
        print("Error leyendo archivo")
        
def actualizar_producto():
    producto = input("Ingresar producto a actualizar:\n")
    existe = False
    for i in comestibles:
        if(producto == i.producto):
            comestibles.remove(i)
            nuevo_producto()
            existe = True
            print(".....Producto Actualizado.....\n")
        else:
            print("Prodcuto no existe en base...!!!\n")

def eliminar_producto():
    producto = input("Ingresar producto a ELIMINAR:\n")
    existe = False
    for i in comestibles:
        if(producto== i.producto):
            comestibles.remove(i)
            existe = True
            print(".....Producto Eliminado.....")
        else:
            print("Producto no Existe...!!!\n")
```
