```python
empresas = []

class Empresa:
    nombre = None
    jefe = None
    telefono = None
    ciudad = None
    correo = None
    
    def __init__ (self, nombre, jefe, telefono, ciudad, correo):
        self.nombre = nombre
        self.jefe = jefe
        self.telefono = telefono
        self.ciudad = ciudad
        self.correo = correo
    
def empresa_nueva():
        nombre = input ("Nombre de Empresa a ingresar:\n")
        jefe = input("Encargado/jefe de la empresa:\n")
        telefono = input("Teléfono de la empresa:\n")
        ciudad = input("Ingrese la ciudad sede de la empresa:\n")
        correo = input("Ingrese correo de la empresa:\n")
        empresas.append(Empresa(nombre, jefe, telefono, ciudad, correo))
        print("Empresa ingresada...!!!")
        
def guardar_empresa():
    try: 
        path = "./Empresas.txt"
        archivo_abierto = open(path, 'w')
        guardar=""
        for nueva_emp in empresas:
            guardar +=nueva_emp.nombre+" - "+nueva_emp.jefe+" - "+nueva_emp.telefono+" - "+nueva_emp.ciudad+" - "+nueva_emp.correo+"\n"
        archivo_abierto.write(guardar)
        archivo_abierto.close()
    except Exception as Error:
        print("Error al guardar empresa...!!!") 
        
def imprimir_empresa():
    try:
        path = "./Empresas.txt"
        archivo_abierto = open(path)
        lista_emp = archivo_abierto.read()
        print(lista_emp)
        archivo_abierto.close()
    except Exception as Error:
        print("Error leyendo Empresas.txt...!!!")
        
def actualizar_empresa():
    nombre = input("Ingrese empresa a actualizar:\n")
    existe = False
    for i in empresas:
        if(nombre == i.nombre):
            empresas.remove(i)
            empresa_nueva()
            existe = True
            print("Empresa Actualizada...!!!")
            break
        if(not existe):
            print("Empresa no Existe...!!!")

def eliminar_empresa():
    nombre = input("Ingresen empresa a eliminar:\n")
    path = "./Empresas.txt"
    infile = open(path, 'r')
    datos = infile.readlines()
    infile.close()
    
    path = "./Empresas.txt"
    outfile = open(path, 'w')
    for i in datos:
        if not i.startswith(nombre):
            outfile.write(i)
    outfile.close()
    print("\nEmpresa Eliminada...!!!\n")
```
